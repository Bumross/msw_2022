### 5. Hledání kořenů rovnice

#### Funkce:

```python
import numpy as np
from numpy import sin, log
import matplotlib.pyplot as plt

f_1 = lambda x : x**2 - 2
f_2 = lambda x : -sin(2*x) + x
f_3 = lambda x : log(x**3) - 1
```

#### Uzavřená metoda - Bisekce, půlení intervalu

```python
f_1 = lambda x : x**2 - 2
f_2 = lambda x : -sin(2*x) + x
f_3 = lambda x : log(x**3) - 1

def uzavrena(funkce, min0, max0, opakovani=100):
    min, max = min0, max0
    for _ in range(1, opakovani):
        koren = (max + min)/2
        if funkce(min)*funkce(max) < 0:
            max = koren
        else:
            min = koren
    return koren

def vysledek(funkce, min, max, koren, nazev):
    print(f"Kořen pro funkci je {koren}")
    print(f"Funkční hodnota v tomto bodě je {funkce(koren)}")
    plt.plot(np.linspace(min, max), funkce(np.linspace(min, max)), "r")
    plt.plot(koren, funkce(koren), "bo")
    plt.title(nazev)
    plt.show()


vysledek(f_1, 0.1, 10, uzavrena(f_1, 0.1, 5), "x**2 - 2")
vysledek(f_2, 0.1, 10, uzavrena(f_2, 0.1, 1), "-sin(2*x) + x")
vysledek(f_3, 0.1, 10, uzavrena(f_3, 1, 3), "log(x**3) - 1")
```

#### Otevřená metoda - Newtonova metoda tečen (pomocí derivace)

```python
import numpy as np
from numpy import sin, log
import matplotlib.pyplot as plt
from scipy.misc import derivative

f_1 = lambda x : x**2-2
f_2 = lambda x : -sin(2*x) + x
f_3 = lambda x : log(x**3) - 1


def otevrena(funkce, bod, opakovani=100):
    for _ in range(1, opakovani):
        koren = bod - (funkce(bod) / derivative(funkce, bod))
        bod = koren
    koren = bod
    return koren


def vysledek(funkce, min, max, koren, nazev):
    print(f"Kořen pro funkci je {koren}")
    print(f"Funkční hodnota v tomto bodě je {funkce(koren)}")
    plt.plot(np.linspace(min, max), funkce(np.linspace(min, max)), "r")
    plt.plot(koren, funkce(koren), "bo")
    plt.title(nazev)
    plt.show()


vysledek(f_1, 0.1, 10, otevrena(f_1, 2), "x**2-2")
vysledek(f_2, 0.1, 10, otevrena(f_2, 0.1), "-sin(2*x) + x")
vysledek(f_3, 0.1, 10, otevrena(f_3, 2), "log(x**3) - 1")
```
