#### Průběh konzumace piva v měsíci

```python
import csv
import numpy as np
import matplotlib.pyplot as plt


date = []
consumed_beer = []

with open("Beer_consumed.csv", "r") as file:
    reader = csv.DictReader(file, delimiter=",")

    for row in list(reader)[120:151]:
        consumed_beer.append(float(row["Beer consumption (liters)"]))
        date.append(row["Date"])

N = len(consumed_beer)
indexes = np.arange(N)+1

plt.bar(indexes, consumed_beer)
plt.xlabel("Květen 2015")
plt.ylabel("Zkonzumované pivo v litrech za den")
plt.title("Zkonzumované pivo v litrech v květnu 2015 v lokaci Sao Paulo, Brazílie, univerzitním kampusu")
plt.show()
```

#### Danceability due popularity of the song

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv("top50.csv")

t_name = data["Track.Name"]
a_name = data["Artist.Name"]

name = tuple(zip(t_name, a_name))
danceability = (data["Danceability"])
bpm = (data["Beats.Per.Minute"])

fig = plt.figure(1, figsize=(20, 20), dpi=50)
ax = fig.add_axes([0.5, 0.5, 1, 1])
fig.autofmt_xdate(rotation=90)

N = np.arange((len(name)))
width = 0.3

plt.bar(N, danceability, width, color="red", edgecolor='black', label='Danceability')
plt.bar(N+width, bpm, width, color="blue", edgecolor='black', label='Beats per Minute')
plt.xticks(N, name)
ax.legend(loc="lower right", fontsize=20)

for label in (ax.get_xticklabels() + ax.get_yticklabels()):
	label.set_fontsize(16)

plt.title("")
plt.show()
```

#### Populární programovací jazyky v roce 2021

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import font_manager as fm
from statistics import mean

data = pd.read_csv("Most Popular Programming Languages from 2004 to 2022.csv")

python = data["Python"][-13:-1:]
c_sharp = data["C#"][-13:-1:]
c_plus = data["C/C++"][-13:-1:]
java = data["Java"][-13:-1:]
    
fig = plt.figure(1, figsize=(10, 10), dpi=50)
ax = fig.add_axes([0.5, 0.5, 1, 1])

plt.title("Populární programovací jazyky v roce 2021")
ax.title.set_fontsize(30)

labels = ["Python", "C#", "C++", "Java"]
fracs = [mean(python), mean(c_sharp), mean(c_plus), mean(java)]
colors = ("yellow", "red", "blue", "green")

patches, texts, autotexts = ax.pie(fracs, colors=colors, labels=labels, autopct="%1.1f%%")

proptease = fm.FontProperties()
proptease.set_size("xx-large")
plt.setp(autotexts, fontproperties=proptease)
plt.setp(texts, fontproperties=proptease)
plt.show()
```

#### Popularita Pythonu

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


data = pd.read_csv("Most Popular Programming Languages from 2004 to 2022.csv")

fig, ax = plt.subplots(figsize=(8, 6))
data["Date"] = pd.to_datetime(data["Date"])
date = data["Date"][::3]
python = data["Python"][::3]

width = 55

ax.bar(date, python, width, color='gray', edgecolor='black', label='Python')

ax.legend(loc="lower right")
plt.title("Procentuální použití jazyku Python od roku 2004 do roku 2022")
plt.show()
```

#### Vývoj populárních jazyků

```python
import matplotlib.pyplot as plt
import numpy as np
import csv
import matplotlib.dates as mdates
import pandas as pd


data = pd.read_csv("Most Popular Programming Languages from 2004 to 2022.csv")

data["Date"] = pd.to_datetime(data["Date"])

date = data["Date"]
python = data["Python"]
java = data["Java"]
c_sharp = data["C#"]
c_plus = data["C/C++"]

fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(date, python, label="Python")
ax.plot(date, java, label="Java")
ax.plot(date, c_plus, label="C++")
ax.plot(date, c_sharp, label="C#")

plt.legend(loc="upper left")
plt.title("Procentuální popularita programovacích jazyků od roku 2004 do roku 2022")
plt.show()
```
