#### Průběh konzumace piva v měsíci

```python
import csv
import numpy as np
import matplotlib.pyplot as plt


date = []
consumed_beer = []

with open("Beer_consumed.csv", "r") as file:
    reader = csv.DictReader(file, delimiter=",")

    for row in list(reader)[120:151]:
            if row["Beer consumption (liters)"] and row["Date"]:
                consumed_beer.append(float(row["Beer consumption (liters)"]))
                date.append(row["Date"])

N = len(consumed_beer)
indexes = np.arange(N)+1

plt.bar(indexes, consumed_beer)
plt.xlabel("Květen 2015")
plt.ylabel("Zkonzumované pivo v litrech za den")
plt.title("Zkonzumované pivo v litrech v květnu 2015 v lokaci Sao Paulo, Brazílie, univerzitní hospodě")

plt.show()
```

#### Závislost konzumace piva na teplotě

```python
import csv
import numpy as np
import matplotlib.pyplot as plt


av_temp = []
consumed_beer = []

with open("Beer_consumed.csv", "r") as file:
    reader = csv.DictReader(file, delimiter=",")

    for row in reader:
            if row["Beer consumption (liters)"] and row["Date"]:
                consumed_beer.append(float(row["Beer consumption (liters)"]))
                av_temp.append(int(float(row["Temperature Average (C)"].replace(",","."))))



consumed_beer = consumed_beer[:]
av_temp = av_temp[:]


plt.bar(av_temp, consumed_beer)
plt.xlabel("Průměrná teplota za den")
plt.ylabel("Zkonzumované pivo v litrech za den")
plt.title("Zkonzumované pivo v litrech v květnu 2015 v lokaci Sao Paulo, Brazílie, univerzitní hospodě")

plt.show()
```

#### Populární programovací jazyky v roce 2021

```python
import csv
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import font_manager as fm
from statistics import mean

python = []
c_sharp = []
c_plus = []
java = []

with open("Most Popular Programming Languages from 2004 to 2022.csv", "r") as file:
    reader = csv.DictReader(file, delimiter=",")

    for row in list(reader)[-13:-1:]:
        python.append(int(float(row["Python"])*100))
        c_sharp.append(int(float(row["C#"])*100))
        c_plus.append(int(float(row["C/C++"])*100))
        java.append(int(float(row["Java"])*100))
    
fig = plt.figure(1, figsize=(10, 10), dpi=50)
ax = fig.add_axes([0.5, 0.5, 1, 1])

plt.title("Populární programovací jazyky v roce 2021")
ax.title.set_fontsize(30)

labels = ["Python", "C#", "C++", "Java"]
fracs = [mean(python), mean(c_sharp), mean(c_plus), mean(java)]
colors = ("yellow", "red", "blue", "green")

patches, texts, autotexts = ax.pie(fracs, colors=colors, labels=labels, autopct="%1.1f%%")

proptease = fm.FontProperties()
proptease.set_size("xx-large")
plt.setp(autotexts, fontproperties=proptease)
plt.setp(texts, fontproperties=proptease)

plt.show()
```

#### Popularita Pythonu

```python
import csv
import numpy as np
import matplotlib.pyplot as plt

year = []
python = []

with open("Most Popular Programming Languages from 2004 to 2022.csv", "r") as file:
    reader = csv.DictReader(file, delimiter=",")
    for row in reader:
        year.append(int(row["Date"][-4::]))
        python.append(int(float(row["Python"])))

width = 0.7

plt.bar(year, python, width, color='gray', edgecolor='black', label='Python')

plt.legend(loc="lower right")
plt.title("Procentuální použití jazyku Python od roku 2004 do roku 2022")

plt.show()
```

#### Vývoj populárních jazyků

```python
import matplotlib.pyplot as plt
import numpy as np

year = []
python = []
c_sharp = []
c_plus = []
java = []

with open("Most Popular Programming Languages from 2004 to 2022.csv", "r") as file:
    reader = csv.DictReader(file, delimiter=",")
    for row in reader:
        year.append(int(row["Date"][-4::]))
        python.append(int(float(row["Python"])))
        c_sharp.append(int(float(row["C#"])))
        c_plus.append(int(float(row["C/C++"])))
        java.append(int(float(row["Java"])))

width = 0.15


plt.bar(year, python, width, color='yellow', edgecolor='black', label='Python')

for i in range(len(year)):
    year[i] += width

plt.bar(year, c_sharp, width, color='red', edgecolor='black', label='C#')

for i in range(len(year)):
    year[i] += width

plt.bar(year, c_plus, width, color='blue', edgecolor='black', label='C++')

for i in range(len(year)):
    year[i] += width

plt.bar(year, java, width, color='green', edgecolor='black', label='Java')
plt.legend(loc="upper left")
plt.title("Vývoj populárních jazyků od roku 2004 do roku 2022")

plt.show()
```
